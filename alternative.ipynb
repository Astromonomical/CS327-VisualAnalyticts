# Alternative: Scatterplot Matrix (GDP, medium income, make ends meet)

### Benefits: all pairs of dimensions are visible
### Drawback: takes up a lot of screen space

```python
import numpy as np
import pandas as pd
import altair as alt
alt.renderers.enable('notebook')
```

```python
PATH = "Data/"
gdp = pd.read_csv(PATH + "gdp_2016.csv")
medianIncome = pd.read_csv(PATH + "median_income_2016.csv") 
makeEndsMeet= pd.read_csv(PATH + "make_ends_meet_2016.csv")

data = gdp.merge(medianIncome,how='inner')
data = data.merge(makeEndsMeet,how='inner')
```

```python
alt.Chart(data).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='country'
).properties(
    width=150,
    height=150
).repeat(
    row=['gdp', 'median_income', 'prct_budget_veryhard', 'prct_budget_hard', 'prct_budget_somehard', 'prct_budget_someeasy', 'prct_budget_easy', 'prct_budget_veryeasy'],
    column=['prct_budget_veryeasy', 'prct_budget_easy', 'prct_budget_someeasy', 'prct_budget_somehard', 'prct_budget_hard', 'prct_budget_veryhard', 'median_income', 'gdp']
).interactive()
```
